```python
import random
from collections import defaultdict
```

```python
names = ["name1", "name2", "name3", "name4", "name5"]
descr = ["descr1", "descr2", "descr3", "descr4"]
labels = ["key1, value1", "key2, value2", "key3, value3", "key4, value4"]
aggregators = ["gauge", "sketch", "exact", "histogram"]
answers = defaultdict(lambda: defaultdict(lambda: [0,0,0,0]))
```

```python
f = open("PrometheusDataSecond.csv", "w")

for _ in range(100000):
    agg = random.choice(aggregators)
    val = random.randint(-50,50)

    nam = names[random.randint(0, 3)]
    des = descr[random.randint(0, 3)]
    lab = random.sample(labels,random.randint(1,4))
    ndl = str(nam) + ", " + str(des) + ", " + str(lab)
    ndl = ndl.replace("[","").replace("]","").replace("'","")
    ndl_srt = str(nam) + ", " + str(des) + ", " + str(sorted(lab))
    ndl_srt = ndl_srt.replace("[","").replace("]","").replace("'","")
    
#     print(ndl)
#     print(ndl_srt)

# TODO: add sketch/exact/histogram records properly to the csv file
    if (agg == "gauge"):
        answers[agg][ndl_srt][0] = val
    elif (agg == "sketch"):
        print(agg)
    elif (agg == "exact"):
        print(agg)
    elif (agg == "histogram"):
        print(agg)
    
#     # mmsc aggregations
#     if (instr == "ivrec" or instr == "ivobs" or instr == "svrec" or instr == "svobs" or instr == "dvrec" or instr == "dvobs" or instr == "fvrec" or instr == "fvobs"):
#         answers[instr][ndl_srt][0] = val if val < answers[instr][ndl_srt][0] else answers[instr][ndl_srt][0]
#         answers[instr][ndl_srt][1] = val if val > answers[instr][ndl_srt][1] else answers[instr][ndl_srt][1]
#         answers[instr][ndl_srt][2] += val
#         answers[instr][ndl_srt][3] += 1
    
#     # counter aggregations
#     else:
#         answers[instr][ndl_srt][0] += val
#         answers[instr][ndl_srt][1] += 1
        
# #     print("{:<6}  {:>6}  {:}".format(agg,val,ndl_srt))

    f.write(agg+','+str(val)+','+"\""+ndl+"\"\n")
    
f.close()
```
